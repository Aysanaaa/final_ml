### Tasks 3 and 4

```python
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']
```

```python
import numpy as np

def normalize_data(X):
    X_norm = (X - np.mean(X))/np.std(X)
    mu = np.mean(X)
    sigma = np.std(X)
    return X_norm, mu, sigma


X, mean, std = normalize_data(X) 
print('Mean: ', mean)
print('Std: ', std)
```

```python
# cost function calculation
import numpy as np

def sigmoid(z):
    return 1/ (1 + np.exp(-z))

def costFunction(theta, X, y):
    m=len(y)
    predictions = sigmoid(np.dot(X,theta))
    error = (-y * np.log(predictions)) - ((1-y)*np.log(1-predictions))
    cost = 1/m * sum(error)
    grad = 1/m * np.dot(X.transpose(),(predictions - y))
    return cost[0] , grad
```

```python
m , n = X.shape[0], X.shape[1]
X= np.append(np.ones((m,1)),X,axis=1)
y=np.array(y).reshape(m,1)
initial_theta = np.zeros((n+1,1))
cost, grad= costFunction(initial_theta,X,y)
print("Cost: ",cost)
# print("Gradient at theta:",grad)
```

```python
# applying linear regression

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = df.drop('SalePrice', axis=1)
y = df['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

model = LinearRegression().fit(X_train,y_train)
y_pred = model.predict(X_test)
y_pred
```

```python
# gradient descent algorithm
from sklearn import metrics
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor().fit(X_train, y_train)

test_pred = sgd_reg.predict(X_test)
train_pred = sgd_reg.predict(X_train)

print('Test set score: ', metrics.r2_score(y_test, test_pred) * 100)
print('Train set score: ', metrics.r2_score(y_train, train_pred) * 100)
```

```python
def gradientDescent(X,y,theta,alpha,num_iters):
    m=len(y)
    J_history =[]
    for i in range(num_iters):
        cost, grad = costFunction(theta,X,y)
        theta = theta - (alpha * grad)
        J_history.append(cost)
    return theta, J_history
```

```python
num_iters = 400 ## Initialize the iteration parameter.
alpha = 0.01 ## Initialize the learning rate.
m , n = X.shape[0], X.shape[1]
X= np.append(np.ones((m,1)),X,axis=1)
y=np.array(y).reshape(m,1)
initial_theta = np.zeros((n+1,1))

theta, J_history = gradientDescent(X,y,initial_theta,alpha,num_iters)
print('New theta: ', theta)
```
